CSC 463 Project One Benchmarking Software

```python
import threading
import time
import matplotlib.pyplot as plt
import numpy as np
```

```python
class customThread(threading.Thread):
  def __init__(self, testType):
    threading.Thread.__init__(self)
    self.testType=testType
    self.count = 0
    self.flag = True

  def run(self):

    if self.testType == 'IOPS':

      while self.flag:
        self.count += 1

    elif self.testType == 'FLOPS':

      while self.flag:
        self.count += 1.0

  def stop(self):
    self.flag = False

  def results(self):
    return(self.count)
```

```python
def timer(timer,threadList):
    time.sleep(timer)

    for i in range(len(threadList)):
      threadList[i].stop()
```

```python
def threadMaker(threadNum, testType):
  threadList = []

  if testType == 'IOPS':

    for i in range(threadNum):
      thread=customThread(testType)
      threadList.append(thread)

  elif testType == 'FLOPS':

    for i in range(threadNum):
      thread = customThread(testType)
      threadList.append(thread)

  return threadList
```

```python
def results(resultsList, testList, testType):

  print(testType, 'Test Results:')

  x = np.array(testList)
  y = np.array(resultsList)

  plt.plot(x, y, marker = 'o')

  plt.xlabel("Number of Threads")
  plt.ylabel("Operations per Second")

  plt.show()
```

```python
def runTest(testList, testType):
  resultsList = []

  if testType == 'IOPS':

    for t in range(len(testList)):
      print(testList[t], 'Thread test')
      total = 0
      threadList=threadMaker(testList[t], testType)
      threadTimer=threading.Thread(target = timer, args = [60, threadList])

      for i in range(len(threadList)):
        threadList[i].start()

      threadTimer.start()

      for j in range(len(threadList)):
        threadList[j].join()

      for l in range(len(threadList)):
        print('Thread', l + 1, ':', threadList[l].results())
        total += threadList[l].results()
      print('IOPS:', total / 60, '\n')
      resultsList.append(total / 60)

  elif testType == 'FLOPS':

    for t in range(len(testList)):
      print(testList[t], 'Thread Test')
      total = 0
      threadList = threadMaker(testList[t], testType)
      threadTimer = threading.Thread(target = timer,args = [60, threadList])

      for i in range(len(threadList)):
        threadList[i].start()

      threadTimer.start()

      for j in range(len(threadList)):
        threadList[j].join()

      for l in range(len(threadList)):
        print('Thread', l + 1, ':', threadList[l].results())
        total += threadList[l].results()
      print('FLOPS:', total / 60, '\n')
      resultsList.append(total / 60)

  results(resultsList, testList, testType)
```

```python
def main():
  print('Welcome to CSC 463 Benchmarking!', '\n')
  testList = []
  flag = True

  while flag:
    tempNum = input('How many threads to test? ')

    while not tempNum.isnumeric() or tempNum == '0':
      print('Please enter a positive whole number')
      tempNum = input('How many threads to test? ')

    testList.append(int(tempNum))

    temp=input('Add more thread tests? (y/n) ')
    while temp.lower() != 'y' and temp.lower() != 'n':
      print('Please type either y or n')
      temp=input('Add more thread tests? (y/n) ')

    if temp.lower() == 'n':
      flag = False

  testNum = input('How many times to run each test? ')
  while not testNum.isnumeric() or testNum == '0':
    print('Please enter a positive whole number')
    testNum=input('How many times to run each test? ')

  testType = input('Which type of test? (IOPS, FLOPS, BOTH) ')
  while testType.upper() != 'IOPS' and testType.upper() != 'FLOPS' and testType.upper() !='BOTH':
    print('Please type either IOPS, FLOPS or BOTH')
    testType = input('Which type of test? (IOPS, FLOPS, BOTH) ')

  print('')
  if testType.upper() == 'BOTH':
    for i in range(int(testNum)):
      print('IOPS Test:', i + 1, '\n')
      runTest(testList, 'IOPS')
      print('FLOPS Test:', i + 1, '\n')
      runTest(testList, 'FLOPS')

  elif testType.upper() == 'IOPS':
    for l in range(int(testNum)):
      print('IOPS Test:', l + 1, '\n')
      runTest(testList, testType.upper())

  elif testType.upper() == 'FLOPS':
    for t in range(int(testNum)):
      print('FLOPS Test:', t + 1, '\n')
      runTest(testList, testType.upper())
```
